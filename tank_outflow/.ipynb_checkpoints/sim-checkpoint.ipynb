```python
from scipy import integrate
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import matplotlib.patches as patches
```

```python
g = 10

x0 = 0
y0 = 0
vx0 = .1
vy0 = 0

X0 = np.array([x0, y0, vx0, vy0])

def F(X, t):
    x = X[0]
    y = X[1]
    vx = X[2]
    vy = X[3]
    
    return [vx,
            vy,
            0,
            -g]

t = np.linspace(0, 15, 1000)
X, infodict = integrate.odeint(F, X0, t, full_output=True)

position_vectors = X[:,0:2]

plt.scatter(position_vectors.T[0], position_vectors.T[1])
plt.show()
```

```python
x0, y0 = [0, 0]

y_gen = lambda x, x0, y0, vx0: -g*(x-x0)**2/(2*vx0**2) + y0

vx0 = 6
y = lambda x: y_gen(x, x0, y0, vx0)

x = np.linspace(0, 20, 1000)
plt.figure(figsize=(4,5))
plt.plot(x, y(x))
plt.xlim([0,20])
plt.ylim([-20,0])
plt.show()
```

```python
x0, y0 = [0, 0]
h0 = 1
A2 = 1
A1 = 100
rho = 1
g = 10


K = A2 / (rho * A1) * np.sqrt(2*g)

h = lambda t: (-K/2 * t + np.sqrt(h0))**2
v0 = lambda t: np.sqrt( 2*g*h(t) )
y_gen = lambda x, x0, y0, vx0: -g*(x-x0)**2/(2*vx0**2) + y0
y = lambda x, t: y_gen(x, x0, y0, v0(t))

#############################################################################

t0 = 0
dt = 0.1
T = 2*np.sqrt(h0)/K
num_of_frames = int(np.floor((T-t0)/dt))
print("number of frames:", num_of_frames)

x = np.linspace(0, 20, 100)

#----------------------------------------------------------------------------

fig = plt.figure()
ax, = plt.plot(x, y(x, t0))

rect_water_pos = (-10, 0)
rect_water_height = 10
rect_water_width = 10
rect_water = patches.Rectangle(rect_water_pos,
                               rect_water_height,
                               rect_water_width,
                               linewidth=1,
                               edgecolor='b',
                               facecolor='b')

rect_container_pos = rect_water_pos

rect_container_height = rect_water_height
rect_container_width = rect_water_width
rect_container = patches.Rectangle(rect_container_pos,
                                   rect_container_height,
                                   rect_container_width,
                                   linewidth=4,
                                   edgecolor='r',
                                   facecolor='none')

plt.gca().add_patch(rect_container)
plt.gca().add_patch(rect_water)

plt.xlim([-11,9])
plt.ylim([-19,11])


def gen():
    frame_number = 0
    
    while frame_number < num_of_frames:
        frame_number += 1
        print(frame_number)
        print("number of frames:", num_of_frames)
        yield frame_number

def animate(frame_number):
    t = t0 + frame_number * dt
    ax.set_ydata(y(x,t))
    new_water_height = rect_water_height * h(t) / h0
    rect_water.set_height(new_water_height)

interval_ms = 50
anim = FuncAnimation(fig, animate, frames=gen, interval=interval_ms, repeat=False)
HTML(anim.to_html5_video())
```
