```python
from scipy import integrate
import numpy as np

import matplotlib.pyplot as plt
```

```python
g = 10

x0 = 0
y0 = 0
vx0 = .1
vy0 = 0

X0 = np.array([x0, y0, vx0, vy0])

def F(X, t):
    x = X[0]
    y = X[1]
    vx = X[2]
    vy = X[3]
    
    return [vx,
            vy,
            0,
            -g]

t = np.linspace(0, 15, 1000)
X, infodict = integrate.odeint(F, X0, t, full_output=True)

position_vectors = X[:,0:2]

plt.scatter(position_vectors.T[0], position_vectors.T[1])
plt.show()
```

```python
x0, y0 = [0, 0]

y_gen = lambda x, x0, y0, vx0: -g*(x-x0)**2/(2*vx0**2) + y0

vx0 = 1
y = lambda x: y_gen(x, x0, y0, vx0)

t = np.linspace(0, 2, 1000)
plt.figure(figsize=(4,5))
plt.plot(t, y(t))
plt.xlim([0,20])
plt.ylim([-20,0])
plt.show()
```
